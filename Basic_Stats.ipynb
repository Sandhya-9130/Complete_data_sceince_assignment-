```python
import pandas as pd

df = pd.read_csv("/content/sales_data_with_discounts.csv")

# Convert 'Date' column to datetime objects if it is not already
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Select only numeric columns before calculating the standard deviation
numeric_df = df.select_dtypes(include=['number'])
std_values = numeric_df.std()

print(std_values)
```
